# Pandas Applying Functions

```python
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset

# Load data
dataset = load_dataset("lukebarousse/data_jobs")
df = dataset["train"].to_pandas()
df_sample = dataset["train"].to_pandas()

# Data cleanup
df["job_posted_date"] = pd.to_datetime(df["job_posted_date"])
df_sample["job_posted_date"] = pd.to_datetime(df_sample["job_posted_date"])
```

## Column apply()

```python
# Calculate project salary next year
df_salary = df[pd.notna(df["salary_year_avg"])].copy()

def projected_salary(salary):
    return salary * 1.03

df_salary["salary_year_inflated"] = df_salary["salary_year_avg"].apply(projected_salary)
df_salary[["salary_year_avg", "salary_year_inflated"]]
```

```python
# Using lambda
df_salary_v2 = df[pd.notna(df["salary_year_avg"])].copy()

df_salary_v2["salary_year_inflated"] = df_salary_v2["salary_year_avg"].apply(lambda salary: salary * 1.03)
df_salary_v2[["salary_year_avg", "salary_year_inflated"]]
```

```python
# Without apply
df_salary_v3 = df[pd.notna(df["salary_year_avg"])].copy()

df_salary_v3["salary_year_inflated"] = df_salary_v3["salary_year_avg"] * 1.03
df_salary_v3[["salary_year_avg", "salary_year_inflated"]]
```

## Converting a list of strings into a list

```python
def clean_list(skill_list):
    # Exclude None values
    if pd.notna(skill_list):
        return ast.literal_eval(skill_list)

df["job_skills"] = df["job_skills"].apply(clean_list)
df["job_skills"]
```

## Row apply()

### Calculate projected salary next year

- Senior roles assume 5%
- Other roles assume 3%

```python
# This one applies to the specified column
df_salary["salary_year_inflated"] = df_salary["salary_year_avg"].apply(lambda salary: salary * 1.03)
```

```python
# This one applies to each row of the specified dataframe
def projected_salary(row):
    if "Senior" in row["job_title_short"]:
        return row["salary_year_avg"] * 1.05
    else:
        return row["salary_year_avg"] * 1.03


df_salary["salary_year_inflated"] = df_salary.apply(projected_salary, axis=1)
df_salary[["job_title_short", "salary_year_avg", "salary_year_inflated"]]
```

### Row apply() with lambda

```python
df_salary_v2 = df_sample[pd.notna(df_sample["salary_year_avg"])].copy()

df_salary_v2["salary_year_inflated"] = (df_salary_v2
                                        .apply(lambda row:  1.05 * row["salary_year_avg"]
                                               if "Senior" in row["job_title_short"]
                                               else 1.03 * row["salary_year_avg"],
                                               axis=1)
                                        )
df_salary_v2[["job_title_short", "salary_year_avg", "salary_year_inflated"]]
```
