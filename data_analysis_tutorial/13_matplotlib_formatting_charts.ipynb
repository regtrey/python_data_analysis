# Matplotlib Formatting Charts

```python
import ast
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset

# Load data
dataset = load_dataset("lukebarousse/data_jobs")
df = dataset["train"].to_pandas()

# Data cleanup
df["job_posted_date"] = pd.to_datetime(df["job_posted_date"])
df["job_skills"] = df["job_skills"].apply(lambda skill: ast.literal_eval(skill)
                                          if pd.notna(skill) else skill)
```

```python
fig, ax = plt.subplots()
```

```python
fig, ax = plt.subplots(2, 2)
```

```python
fig, ax = plt.subplots(1, 2)

df["job_title_short"].value_counts().plot(kind="bar", ax=ax[1])
df["job_schedule_type"].value_counts().head(3).plot(kind="bar", ax=ax[0])
```

```python
fig, ax = plt.subplots(1, 2)

df["job_title_short"].value_counts().plot(kind="bar", ax=ax[1])
df["job_schedule_type"].value_counts().head(3).plot(kind="bar", ax=ax[0])

fig.tight_layout()
```

```python
fig, ax = plt.subplots(1, 2)

ax[0].bar(df["job_schedule_type"].value_counts().head(3).index, df["job_schedule_type"].value_counts().head(3))
ax[1].bar(df["job_title_short"].value_counts().index, df["job_title_short"].value_counts())

plt.tight_layout()
```

## Example - Counts of Top Skills in Job Postings

```python
df_skills = df.copy()
df_skills = df_skills.explode("job_skills")
skills_count = df_skills.groupby(["job_skills", "job_title_short"]).size()
df_skills_count = skills_count.reset_index(name="skill_count")
df_skills_count.sort_values(by="skill_count", ascending=False, inplace=True)
df_skills_count
```

```python
job_titles = ["Data Scientist", "Data Engineer", "Data Analyst"]
```

```python
# With enumerate it provides the list's index and value
for i, job_title in enumerate(job_titles):
    print(i, job_title)
```

```python
fig, ax = plt.subplots(3, 1)

# With enumerate it provides the list's index and value
for i, job_title in enumerate(job_titles):
    print(df_skills_count[df_skills_count["job_title_short"] == job_title].head(5))
```

```python
fig, ax = plt.subplots(3, 1)

# With enumerate it provides the list's index and value
for i, job_title in enumerate(job_titles):
    df_plot = df_skills_count[df_skills_count["job_title_short"] == job_title].head(5)
    df_plot.plot(kind="barh", x="job_skills", y="skill_count", ax=ax[i], title=job_title)
    ax[i].invert_yaxis()
    ax[i].set_ylabel("")
    ax[i].legend().set_visible(False)
    ax[i].set_xlim(0, 120000)

fig.suptitle("Counts of Top Skills in Job Postings", fontsize=15)
fig.tight_layout()
```
