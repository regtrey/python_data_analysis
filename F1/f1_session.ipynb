```python
td = pd.Timedelta(minutes=1, seconds=10, milliseconds=250)
td
```

```python
# Converts into nanoseconds
# 1 second = 1,000,000,000 nanoseconds
td.to_timedelta64()
```

```python
td.to_timedelta64() / 1000000000
```

```python
td64 = td.to_timedelta64()
td64
```

```python
td64.astype("timedelta64[s]").astype(float)
```

```python
td64.astype("timedelta64[ms]").astype(int)
```

```python
# 1s = 1,000ms
# 60s = 60,000ms
minutes = td64.astype("timedelta64[ms]").astype(int) // 60000
minutes
```

```python
# 1s = 1,000ms
# 60s = 60,000ms
# Here we get the remainder of the total ms / ms per minute
# Then divide it by 1000 ms which is equals to a second
# Double // division to round it and separate the milliseconds
seconds = (td64.astype("timedelta64[ms]").astype(float) % 60000) / 1000
seconds
```

```python
milliseconds = td64.astype("timedelta64[ms]").astype(int) % 1000
milliseconds
```

```python
laptime = f"{minutes}:{seconds}:{milliseconds}"
laptime
```

```python
milliseconds + milliseconds
```

### Formatting y-axis into a laptime string

```python
time_y = np.array([70.325, 70.390, 70.132, 70.160, 70.542])
lap_x = np.array([2, 4, 6, 8, 10])

sample_df = pd.DataFrame({
    "time": time_y,
    "lap": lap_x
})

fig, ax = plt.subplots()

sample_df.plot(kind="line", x="lap", y="time", ax=ax)

def laptime_format(y, pos):
    minutes = int(y // 60)
    seconds = y % 60
    return f"{minutes:02d}:{seconds:06.3f}"

ax.yaxis.set_major_formatter(plt.FuncFormatter(laptime_format))
```

```python
time_y = np.array([70.325, 70.390, 70.132, 70.160, 70.542])
lap_x = np.array([2, 4, 6, 8, 10])

sample_df = pd.DataFrame({
    "time": time_y,
    "lap": lap_x
})

fig, ax = plt.subplots()

sample_df.plot(kind="line", x="lap", y="time", ax=ax)

def laptime_format(y, pos):
    minutes = int(y // 60)
    seconds = y % 60
    return f"{minutes:02d}:{seconds:06.3f}"

ax.yaxis.set_major_formatter(plt.FuncFormatter(laptime_format))
ax.set_ylim(sample_df['time'].min() - 0.5, sample_df['time'].max() + 0.5)
```

## Plot qualifying times

```python
session = fastf1.get_session(2025, "Azerbaijan Grand Prix", "Q")
session.load()
session.results
```

```python
# Make a copy first of the session results' dataframe
q1 = session.results.copy()

# Make a column for the converted timedelta into seconds
q1["total_seconds"] = session.results["Q1"].dt.total_seconds()
q1["total_seconds"]
```

```python
q1
```

```python
q1.plot(kind="bar", x="Abbreviation", y="total_seconds")
plt.ylim(99.5, 105.5)
plt.xticks(rotation=45)
plt.show()
```

```python
fig, ax = plt.subplots()

q1.plot(kind="bar", x="Abbreviation", y="total_seconds", ax=ax)

def lap_formatter(y, pos):
    minutes = int(y // 60)
    seconds = y % 60
    return f"{minutes:02d}:{seconds:06.3f}"

ax.yaxis.set_major_formatter(plt.FuncFormatter(lap_formatter))

plt.ylim(99.5, 105.5)
plt.xticks(rotation=45)
plt.show()
```
